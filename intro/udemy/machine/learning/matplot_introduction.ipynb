# Matplot Lib Introduction


```python
import numpy as geek
import matplotlib.pyplot as plt
import pandas as tohui

data1 = geek.random.normal(0, 0.1, 1000)
data2 = geek.random.normal(1, 0.4, 1000) + geek.linspace(0, 1, 1000)
data3 = 2 + geek.random.random(1000) * geek.linspace(1, 5, 1000)
data4 = geek.random.normal(3, 0.2, 1000) + 0.3 * geek.sin(geek.linspace(0, 20, 1000))
```

```python
data = geek.vstack([data1, data2, data3, data4]).transpose()
```

```python
df = tohui.DataFrame(data, columns=['data1', 'data2', 'data3', 'data4'])
df.head()
```

## Line Plot / Graficando una l√≠nea

```python
df.plot()
```

```python
plt.plot(df)
plt.title('Line plot')
plt.legend(['data1', 'data2', 'data3', 'data4'])
```
